##### Implement K-Means clustering on Iris.csv dataset. Determine the number of clusters
##### using the elbow method.
##### Dataset Link: https://www.kaggle.com/datasets/uciml/iris

```python
# Necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.datasets import load_iris
data = load_iris().data
target = load_iris().target
col_names = load_iris().feature_names
df = pd.concat([pd.DataFrame(data,columns=col_names),pd.DataFrame(target,columns=['target'])],axis = 1)
df.head()
```

```python
df.shape
```

```python
X,y = df.drop(columns=['target']),df['target']
```

```python
sc = StandardScaler()
sc.fit_transform(X)
```

```python
sse = []
for i in range (1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(10,10))
plt.plot(range(1,10),sse,marker = 'o')
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('sse')
plt.show()
```

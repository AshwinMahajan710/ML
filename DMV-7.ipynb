```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
# 1 > load all the data
data_csv = pd.read_csv('customers.csv')
data_excel = pd.read_excel('customers.xlsx')
data_json = pd.read_json('customers.json')
```

```python
# 2 . Perform data cleaning operations, such as handling missing values, removing duplicates, or correcting inconsistencies.
```

```python
data_csv.duplicated(subset=['first_name','last_name','email']).sum()
```

```python
print(data_csv.isnull().sum())
print(data_excel.isnull().sum())
print(data_json.isnull().sum())
```

```python
print(data_csv.info())
print(data_json.info())
print(data_excel.info())
```

```python
# 3 . Convert the data into a unified format, such as a common dataframe or data structure, to enable seamless analysis.
data_combine = pd.concat([data_csv,data_excel,data_json],axis = 0)
data_combine.shape
```

```python
data_combine = data_combine.drop_duplicates(subset=['first_name','last_name','email'])
data_combine.shape
```

```python
# 4. Perform data transformation tasks, such as merging multiple datasets, splitting columns, or deriving new variables.
```

```python
data_combine['full_name'] = data_combine['first_name'] + ' ' + data_combine['last_name']
data_combine.drop(columns = ['first_name','last_name'], inplace = True)
data_combine.head()
```

```python
# 6. Analyze the sales data by performing descriptive statistics, aggregating data by specific variables, or calculating
#    metrics such as total sales, average order value, or product category distribution.
```

```python
data_combine.describe()
```

```python
gb = data_combine.groupby('job').agg({
    'orders': 'sum',
    'spent' : 'mean'
})
gb
```

```python
# 7. Visualization
```

```python
sns.barplot(x = 'job',y='spent',data=data_combine,hue = 'gender')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.pie(data_combine['gender'].value_counts(),labels = data_combine['gender'].value_counts().index,)
plt.show()
```
